```python
# Import libraries
from swolfpy_processmodels import LF
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set font for plots
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "14"

# Set specs for pandas
pd.set_option('display.max_colwidth', 0)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.precision', 2)
```

```python
model = LF()
model.calc()
```

```python
model.lcht_Qlty.iloc[:, [0, 3, 4, 5, 6, 7]]
```

```python
model._param3[['Generated Leachate (m3/Mg MSW)']].head()
```

```python
model.lcht_conc.head()
```

```python
mass_lcht_emission = pd.DataFrame((model.lcht_conc.values.T * model._param3['Generated Leachate (m3/Mg MSW)'].values).T * 1000,
             columns=model.lcht_conc.columns)
mass_lcht_emission.head()
```

```python
pd.DataFrame(mass_lcht_emission.sum(), columns=['Mass in 100 years leachate (kg)'])
```

```python
def plot_cumulative(compound):
    cumlative = []
    j=0
    for i in mass_lcht_emission[compound]:
        j += i
        cumlative.append(j)
    fig, ax = plt.subplots(1, 1, figsize=(8, 5))
    ax.scatter(np.arange(1, 101), cumlative, label=compound, s=4)
    ax.set_xlabel('years after waste buried')
    ax.set_ylabel(f'{compound} (kg)')
    ax.axhline(0.95 * j, c='r', label='95% of total release')
    ax.axhline(0.99 * j, c='g', label='99% of total release')
    ax.legend()
```

```python
plot_cumulative('BOD5, Biological Oxygen Demand')
```

```python
plot_cumulative('COD, Chemical Oxygen Demand')
```

```python
plot_cumulative('Phosphate')
```

```python
plot_cumulative('TSS')
```

```python
plot_cumulative('Barium')
```
