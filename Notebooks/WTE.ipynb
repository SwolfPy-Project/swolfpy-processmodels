# Waste to Energy (WTE)

```python
# Import libraries
from swolfpy_processmodels import WTE
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set font for plots
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "12"

# Set specs for pandas
pd.set_option('display.max_colwidth', 0)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
model = WTE()
model.calc()
report = model.report()
```

## Energy Balance

```python
Energy = pd.DataFrame(index=[x.replace('_', ' ') for x in model.Index])

Energy['Produced Electricity'] =  model.Energy_Calculations['Energy_Recovered_as_Electricity'].values
Energy['Used Electricity'] = -model.Energy_Calculations['Electricity_Use'].values
Energy['Heat Recovered'] = model.Energy_Calculations['Heat_Recovered'].values / 3.6

fig, ax=plt.subplots(figsize=(15, 4))

Energy.plot(kind='bar', stacked=True, ax=ax)
```

## Waste Products

```python
Waste_products = pd.DataFrame(report['Waste']).transpose()
Waste_products
```

```python
fig, ax = plt.subplots(figsize=(15, 4))
Waste_products.plot.bar(ax=ax)
ax.set_ylim((0, 1))
```
